```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
# reading ImdbTitleBasics.csv
title_file = "Resources/ImdbTitleBasics.csv"
title_df = pd.read_csv(title_file)
```

```python
# Identifying the placeholder for "\\N" value
title_df.tail(1)
```

```python
# Replacing  "\\N" value with Nan
title_df=title_df.replace(title_df.tail(1)["endYear"].to_list()[0],np.nan)
```

```python
# reading ImdbTitleRatings.csv
ratings_file = "Resources/ImdbTitleRatings.csv"
ratings_df = pd.read_csv(ratings_file)
```

```python
# Removing the first two symbols from the tconst column in order to use it as a integer PK
title_df['tconst'] = title_df['tconst'].str[2:]
title_df
```

```python
# converting the tconst to integer
title_df['tconst'] = title_df['tconst'].astype(int)

title_df
```

```python
# converting the tconst to integer
ratings_df['tconst']=ratings_df['tconst'].str[2:].astype(int)
ratings_df
```

```python
# Renaming the columns
title_df_renamed = title_df.rename(columns={"tconst":"id",'titleType':'type','primaryTitle':'primarytitle','originalTitle':'originaltitle','startYear':'year','runtimeMinutes':'runtime' })
title_df_renamed
```

```python
# Renaming the columns
ratings_df_clean = ratings_df.rename(columns={"tconst":"id", 'averageRating':'ratings','numVotes':'votes'})
ratings_df_clean
```

```python
# Dropping two columns
title_df_clean = title_df_renamed.drop(columns=["endYear", "isAdult"])
title_df_clean
```

```python
# Dropping any duplicates with the same id
title_df_clean=title_df_clean.drop_duplicates(subset=['id'])
```

```python
# Replacing all the NaN values with 0 and converting the year and runtime columns to integer
title_df_clean = title_df_clean.replace(np.nan, 0)
title_df_clean['year']=title_df_clean['year'].astype(int)
title_df_clean['runtime']=title_df_clean['runtime'].astype(int)
title_df_clean
```

```python
# Dropping any duplicates with the same id
ratings_df_clean=ratings_df_clean.drop_duplicates(subset=['id'])
```

```python
# Setting the id column as index
title_df_clean.set_index("id", inplace=True)
title_df_clean
```

```python
# Setting the id column as index
ratings_df_clean.set_index("id", inplace=True)
ratings_df_clean
```

```python
# Establishing a connection with postgres
connection_string = "postgres:postgres@localhost:5432/imdb_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# checking table's names
engine.table_names()
```

```python
# Load DataFrames into database
title_df_clean.to_sql(name='imdbtitlebasics', con=engine, if_exists='append', index=True)
```

```python
ratings_df_clean.to_sql(name='imdbtitleratings', con=engine, if_exists='append', index=True)
```

```python
# Confirm data has been added by querying the customer_location table
pd.read_sql_query('select * from imdbtitleratings', con=engine).head()
```
